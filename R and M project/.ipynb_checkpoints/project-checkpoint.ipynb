```python
import pandas as pd
import json
import requests
```

```python

####### URL-INFO ###################################
url_base = "https://rickandmortyapi.com/api"
url_characters = "https://rickandmortyapi.com/api/character"
url_locations ="https://rickandmortyapi.com/api/location"
url_episodes = "https://rickandmortyapi.com/api/episode"
characters_alive = url_characters + "?status=Alive"
characters_dead = url_characters + "?status=Dead"


####### functions ###################################

def fetch_char_data(characters_alive, characters_dead):
    data_characters_alive=requests.get(characters_alive).json()
    df_characters_alive= pd.json_normalize(data_characters_alive["results"])
    data_characters_dead=requests.get(characters_dead).json()
    df_characters_dead= pd.json_normalize(data_characters_dead["results"])
    df_characters = pd.concat([df_characters_alive,df_characters_dead], ignore_index= True)
    return df_characters


def location_char_data () :
    data_characters_location=requests.get(url_locations).json()
    df_characters_loc= pd.json_normalize(data_characters_location["results"])  
    return df_characters_loc


def episode_char_data () :
    data_characters_episode=requests.get(url_episodes).json()
    df_characters_episode= pd.json_normalize(data_characters_episode["results"])  
    return df_characters_episode

def last_char_episode (strings) :
    new_list =[]
    for string in strings :
        new_list.append(string.rsplit("/" ,1)[-1])
    return list(new_list)

fetch_char_data(characters_alive, characters_dead)





```

```python
df_a = fetch_char_data(characters_alive, characters_dead)
df_a["len_episode"] = df_a["episode"].apply(lambda x : len(x) )
df_a["episode_num"] = df_a["episode"].apply(lambda x : last_char_episode(x))
df_a = df_a.drop(axis = 1 , columns = ["image","origin.url","origin.name","location.url","url","episode"])
df_a_filter = df_a[(~df_a["name"].isin(["Morty Smith","Summer Smith","Beth Smith","Jerry Smith","Rick Sanchez"]))
             & (df_a["status"] == "Alive")]
df_a_filter[df_a_filter["len_episode"] == df_a_filter["len_episode"].max()]






```

```python
df_b = episode_char_data ()
df_b["Season"] = df_b["episode"].str[:3]
df_b["Episde"] = df_b["episode"].str[3:]
df_b.drop(axis = 1 , columns = ["characters","url", "created","episode"] , inplace= True)
season = df_b.groupby("Season")
first = season.first()
last = season.last()
first_last = pd.concat([first,last], ignore_index= True)
first_last["air_date"] = pd.to_datetime(first_last["air_date"])
first_last.sort_values(by  = "air_date")
```

```python
df_c = location_char_data ()
df_c["list of residents"] = df_c["residents"].apply(lambda x : last_char_episode(x))
df_c["number of residents"] = df_c["residents"].apply(lambda x : len(x))
df_c.drop(axis= 1 ,  columns = ["type","dimension", "url","created","residents", "list of residents"], inplace= True)
df_c.nlargest(1 , columns ="number of residents")
```

```python
df_d =  season.count()
df_d.drop(axis = 1 , columns= ["id","name","air_date"], inplace= True )
df_d.rename(columns= {"Episde" :  "Number of episdes in each season"} , inplace= True)
df_d
```

```python
df_e = df_b
df_e["air_date"] = pd.to_datetime(df_e["air_date"])
df_e["next air date"]= df_e["air_date"].shift(periods= -1 , axis = 0  )
df_e =df_e.drop(axis =1, columns= ["id","name","Season", "Episde"] )
df_e["day diff"] = df_e["next air date"] - df_e["air_date"]
df_e_sort = df_e.sort_values("day diff" ,ascending= False)
df_e_nat=df_e_sort.dropna()
df_e_nat.nlargest(1, columns = "day diff")
```
